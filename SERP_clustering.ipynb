```python
sites={}
for key in final_dictionnary:
    for website in final_dictionnary[key]:
        if(website not in sites):
            sites[website]=[key]
        else : 
            sites[website].append(key)    
len(sites)          
```

```python
sites
```

```python
table_websites=[]

for key in sites :
    table_websites.append(key) 
table_websites
```

```python
raw_X={}
for key in final_dictionnary:
    for item in final_dictionnary[key]:
        if(key not in raw_X):
            raw_X[key]=[table_websites.index(item)]
        raw_X[key].append(table_websites.index(item)) 
raw_X           
```

```python
len(raw_X)
```

```python
for item in raw_X : 
     if len(raw_X[item]) > 8:
        raw_X[item]= raw_X[item][:8]
raw_X        
```

```python
import numpy as np
# Define the universe of integers you care about
integer_universe = set(range(1,len(table_websites)))  # Example: integers from 1 to 100

# Transform raw_X into binary feature vectors
X_transformed = []
for item in raw_X:
    binary_vector = [1 if i in raw_X[item] else 0 for i in integer_universe]
    X_transformed.append(binary_vector)

# Convert to numpy array
X_transformed = np.array(X_transformed)
```

```python
from sklearn.cluster import KMeans
import numpy as np 


num_clusters=12
#X=[]
#for item in raw_X :
    #X.append(raw_X[item])
#X = np.array(X)
X= X_transformed  
kmeans = KMeans(n_clusters=num_clusters)
kmeans.fit(X)

# Get the cluster labels
labels = kmeans.labels_
```

```python
import matplotlib.pyplot as plt
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Visualize the clusters
plt.figure(figsize=(8, 6))

# Plot data points with color-coded clusters
for i in range(num_clusters):
    cluster_points = X[np.where(labels == i)]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], label=f'Cluster {i+1}')

# Plot centroids
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', color='red', s=100, label='Centroids')

plt.title('KMeans Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()
```
